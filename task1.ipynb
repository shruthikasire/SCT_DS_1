```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("population_by_age_group.csv")

df.columns = df.columns.str.strip()

country_name = "India"
country_data = df[df["Country"] == country_name].iloc[0]

# Correct column names from YOUR dataset
age_columns = ['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years']

population_values = [country_data[col] for col in age_columns]

# bar chart
plt.figure(figsize=(9,5))
plt.bar(age_columns, population_values)

plt.xlabel("Age Groups")
plt.ylabel("Population")
plt.title(f"Population Distribution by Age Group in {country_name}")
for i, val in enumerate(population_values):
    plt.text(i, val, f"{val:,}", ha='center', va='bottom')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['25-64 years'], bins=20)

plt.xlabel("Population (2564 years)")
plt.ylabel("Number of Countries")
plt.title("Distribution of 2564 Years Population Across Countries (2021)")

plt.show()
```
